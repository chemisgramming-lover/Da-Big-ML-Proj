### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.insert(0, '..')

from src.data_loader import load_cifar10, get_all_class_names
from src.utils import plot_class_distribution

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

### Load Dataset

```python
# Load CIFAR-10
X_train, y_train, X_test, y_test = load_cifar10()

print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
print(f"Number of classes: {len(np.unique(y_train))}")
print(f"Pixel value range: [{X_train.min():.2f}, {X_train.max():.2f}]")
```

### Dataset Statistics

```python
# Class distribution
class_names = get_all_class_names()
unique_train, counts_train = np.unique(y_train, return_counts=True)

df_dist = pd.DataFrame({
    'Class': [class_names[i] for i in unique_train],
    'Count': counts_train,
    'Percentage': (counts_train / len(y_train) * 100).round(2)
})

print("\nTraining Set Class Distribution:")
print(df_dist)

# Statistical summary
print(f"\n\nPixel Statistics:")
print(f"Mean: {X_train.mean():.4f}")
print(f"Std:  {X_train.std():.4f}")
print(f"Min:  {X_train.min():.4f}")
print(f"Max:  {X_train.max():.4f}")
```

### Visualize Sample Images

```python
# Plot sample images
fig, axes = plt.subplots(4, 8, figsize=(14, 8))
axes = axes.ravel()

for i, ax in enumerate(axes):
    idx = np.random.randint(0, len(X_train))
    ax.imshow(X_train[idx])
    ax.set_title(class_names[y_train[idx]], fontsize=10)
    ax.axis('off')

plt.suptitle('Random CIFAR-10 Training Samples', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Class Distribution Visualization

```python
plot_class_distribution(y_train, y_test)
```

### Channel Analysis

```python
# Analyze each channel
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
channel_names = ['Red', 'Green', 'Blue']

for i, (ax, name) in enumerate(zip(axes, channel_names)):
    ax.hist(X_train[:, :, :, i].ravel(), bins=50, alpha=0.7, color=['red', 'green', 'blue'][i])
    ax.set_title(f'{name} Channel Distribution')
    ax.set_xlabel('Pixel Value')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Observations & Insights

1. **Balanced Dataset**: CIFAR-10 has well-balanced class distribution (5000 samples per class)
2. **Image Quality**: 32x32 low-resolution images - challenging for complex models
3. **Color Distribution**: All channels have similar distributions
4. **Data Augmentation**: Essential due to small image size and limited training samples
